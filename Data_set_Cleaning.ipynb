```python
data_set['Easy Apply'].unique()
```

```python
# Replace specific terms in 'Easy Apply' column
data_set['Easy Apply'].replace({'-1': 'FALSE'}, inplace=True)
```

```python
data_set['Easy Apply'].unique()
```

```python
data_set['Established'].unique()
```

```python
# Replace specific terms in 'Established' column
data_set['Established'].replace({-1: 'unknown'}, inplace=True)
```

```python
data_set['Established'].unique()
```

```python
data_set['Salary'].unique()
```

```python
# Replace specific terms in 'Salary' column
data_set['Salary']=data_set['Salary'].str.replace('k', '000')
```

```python
data_set['Salary']
```

```python
data_set['Salary']=data_set['Salary'].str.replace('$', '')
```

```python
data_set['Salary']
```

```python
#Age
avg_age = data_set['Age'].mean()
avg_age
```

```python
# fill the missing values with mean
data_set['Age'] = data_set.Age.fillna(avg_age)
data_set['Age'] = data_set.Age.round(decimals=1)
```

```python
#Rating
r1=data_set['Rating'].mean()
r1
```

```python
# fill the missing values with mean
data_set['Rating'] = data_set.Rating.fillna(r1)
data_set['Rating'] = data_set.Rating.round(decimals=0)
```

```python
# Replace specific terms in 'Rating' column
data_set['Rating'].replace({-1.0: r1}, inplace=True)
```

```python
data_set['Rating']
```

```python
data_set
```

```python
data_set['Location'].value_counts()
```

```python
data_set['Location'] = data_set['Location'].replace({'New York,Ny':'New York', 'India,In':'India','India In':'India', 'Australia Aus':'Australia'})
```
